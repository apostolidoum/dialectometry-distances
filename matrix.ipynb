Gemerate a matrix for each metric (upos, ufeats, uas)

````python
import os
import json
import pandas as pd
````

````python
def get_languages(folder_path='metrics'):
    """
    Parses the metrics folder to extract a unique, sorted list of 
    all source and target languages found in the filenames.
    """
    languages = set()
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            # Remove suffix and split by '-on-'
            name_part = filename.replace('-ud-test.json', '')
            parts = name_part.split('-on-')
            if len(parts) == 2:
                languages.add(parts[0])
                languages.add(parts[1])
    
    # Return sorted list to ensure consistent matrix order
    return sorted(list(languages))
````

````python
def create_single_metric_matrix(metric_name, folder_path='metrics'):
    languages = get_languages(folder_path)
    
    # Initialize the empty DataFrame with the extracted languages
    matrix = pd.DataFrame(index=languages, columns=languages)
    
    # Iterate through files and fill the DataFrame
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            name_part = filename.replace('-ud-test.json', '')
            source, target = name_part.split('-on-')
            
            with open(os.path.join(folder_path, filename), 'r') as f:
                data = json.load(f)
                
                if metric_name in data:
                    matrix.loc[source, target] = data[metric_name]
    
    return matrix
````

````python
upos_df = create_single_metric_matrix('upos')
print("--- UPOS Matrix ---")
print(upos_df)
````

````python
ufeats_df = create_single_metric_matrix('ufeats')
print("--- UFeats Matrix ---")
print(ufeats_df)
````

````python
uas_df = create_single_metric_matrix("uas")
print("--- UAS Matrix ---")
print(uas_df)
````

Plot the matrices as heatmaps

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
def plot_heatmap(matrix, metric_name, filename):
    """
    Plots a heatmap for the given matrix and saves it as a PNG file.
    """
    plt.figure(figsize=(8, 8))
    # We ensure the matrix is float type for plotting
    sns.heatmap(matrix.astype(float), annot=True, cmap='viridis', fmt='.2f')
    plt.title(f'Heatmap for {metric_name.upper()}')
    plt.tight_layout()
    plt.savefig(filename)
    plt.close()
````

````python
plot_heatmap(upos_df, "upos", "upos_heatmap.png")
````

````python
plot_heatmap(ufeats_df, "ufeats", "ufeats_heatmap.png")
````

````python
plot_heatmap(uas_df, "uas", "uas_heatmap.png")
````
